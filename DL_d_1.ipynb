NUMY-NUMERICAL PYTHON

```python
#Import Numpy
import numpy as np

list = [1,2,3,4]
arr = np.array([1,2,3,4])

print(list*2) #repeate the same list
print(arr*2) #multplies each element
#numpy is faaster + smarter + math-ready
```

```python
#multi-dimentional arrays
#single dimention
arr = np.array([10, 20, 30, 40, 50])
print(arr)
#two dimentional array---rows x columns
arr2 = np.array([[1,2,3],[4,5,6]])

print(arr2)
```

```python
#pre-filled array
z = np.zeros(5)
o = np.ones(5)
print(z)
print(o)
r = np.arange(1,11)
print(r)
```

```python
#How to generate a random number
nums = np.random.rand(5)
print(nums)
```

```python
#array properties
a = np.array([[1,2,3],[4,5,6]]) #2d array
print(a.shape) #rows and column
print(a.size) #total number of elements
print(a.ndim) #to check the dimentionality
```

```python
#indexing and slicing in arrays
a = np.array([10, 20, 30, 40])
print(a[-2]) #indexing
print(a[1:4]) #slicing

#2D indexing
b = np.array([[1, 2, 4],
              [5, 6, 7]])
print(b[0,1])
print(b[1,2])
```

```python
#Mathematical operations on array
#element wise match
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a+b)
print(a*b)
```

```python
#Statistical function
a = np.array([10, 20, 30, 40, 50])

print(np.mean(a))
print(np.max(a))
print(np.min(a))
print(np.std(a))
```

```python
#universal function
a = np.array([1,4,9])

print(np.sqrt(a))
print(np.log(a))
print(a + 10) #added each element without using loop
```

```python
#reshaping in array
a = np.arange(1,13)
print(a)
#reshape changes 1D into 2D
b = a.reshape(3, 4)
print(b)
#flatten changes back 2D into 1D
print(b.flatten())
```

```python
#Matrix operations // simulation how neurons calculate output
a = np.array([[1,2,],
             [3,4]])
b = np.array ([[5,6],
              [7,8]])

print(np.dot(a,b))
#transpose
print(a.T)
```

```python
#boolean Masking(Filtering)
a = np.array([10, 20, 30, 40, 50])

print(a[a >= 20])
```

```python
#saving and loading
np.save("data.npy",a)
b = np.load("data.npy")
print(b)
```
